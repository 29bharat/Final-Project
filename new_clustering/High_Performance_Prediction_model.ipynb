To predict distict with hight graduation rates based on district funds financingn we will use deep learning model. As we split districts on two groups: high permormance (graduation rade >75%) and low performance (graduation rates <75%) it's assumed that sigmoid activation function will give the highest accuracy

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
```

Previously in preprocessing we have already made feature standardization and encoding, so we will use dataframe from df_encoded

```python
# Import our input dataset
df = pd.read_csv('Sources/df_encoded.csv')
df.head()
```

```python
#Drop column Unnamed: 0
df = df.drop(columns = ['Unnamed: 0'])
df
```

```python
#split Fin_Amount into bins
groups = [0, 500000, 1500000, 3000000, 10000000, 482792753.94000006]
new_labels = ["<500k", "500k-1500k", "1500k-3000k", "3000k-10000k", ">10000k"]
df["Amount"] = pd.cut(df["Amount"], groups, labels = new_labels)
df.head()
```

```python
#Encode Fin_Amount
df = pd.get_dummies(df, columns = ['Amount'])
df
```

```python
#Drop Unnecessary columns
df = df.drop(columns = ['CDS_CODE', 'Year', 'ETHNIC', 'GR_12', 'ENR_TOTAL', 'GRADS', 'UC_GRADS', 'District_Code'])
df
```

```python
#Recode graduation rate
rates = ["<25%", "25%-50%", "50%-75%", ">75%"]
replaces = ['0', '0', '0', '1']
for i in range (0,4):
    df.loc[df.GR_RATE == rates[i], 'GR_RATE'] = replaces[i]
df
```

```python
df['GR_RATE'] = df[['GR_RATE']].astype(int)
df.dtypes
```

```python
#Split the data into features and target arrays
y = df['GR_RATE'].values
X = df.drop(['GR_RATE'],1).values
```

```python
#Split the data into train ans test model
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 50)
```

```python
#Standardize the numerical features
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

Build a Deep Learning Model -- number of nodes in hidden layers, number of layers, activation functions, and number of epochs might be changed

```python
#Define the model 
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 16
hidden_nodes_layer2 = 8
hidden_nodes_layer3 = 3
nn = tf.keras.models.Sequential()
```

```python
#First Layer
nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer1,
                     input_dim = number_input_features, activation ="relu"))
```

```python
#Second Layer
nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer2, activation ="relu"))
```

```python
nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer3, activation ="relu"))
```

```python
#Output layer
nn.add(tf.keras.layers.Dense(units = 1, activation ="sigmoid"))
```

```python
#Check summary
nn.summary()
```

```python
#Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train,y_train,epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
