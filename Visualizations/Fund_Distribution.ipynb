```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
```

```python
# Files to load
finance_data_load = "../Resources/Finance Data/School Budget Data '16-'17.csv"
```

```python
# Store data into Pandas DataFrame
finance_data_df = pd.read_csv(finance_data_load)
finance_data_df.head(10)
```

```python
# Change column names
finance_data_df.columns = ['Ccode', 'Dcode', 'SchoolCode', 'Fiscalyear', 'Period','Colcode','Account','Fund','Resource','Projectyear','Goal','Function','Object','Value']
finance_data_df.head(10)
```

```python
# Add original column name back as a row
add_row = pd.DataFrame({'Ccode': 1, 'Dcode': 10017, 'SchoolCode': 0, 'Fiscalyear': 2016, 'Period': 'A','Colcode':'BA',
                        'Account':'0100000000000008011','Fund': 1,'Resource': 0,'Projectyear': 0,'Goal': '0000','Function': 0,
                        'Object': 8011,'Value': 3968003.13}, index=[0])
finance_data_df = pd.concat([add_row, finance_data_df]).reset_index(drop = True)
finance_data_df.head(10)
```

```python
# Delete values with Function value 0
finance_data_df = finance_data_df[finance_data_df.Function != 0]
finance_data_df
```

```python
# Replace Function Codes with groups Instruction, Student Services, Extra-Curricular/Community, Adminsitration, Operations
finance_data_df['Function'] = finance_data_df['Function'].replace([1110,1120,1130,1180,1190,2200,3110,3120,3130,3140,3150,3160,
                                                                   3600,3700,3900,4000], 'Student Services')

finance_data_df['Function'] = finance_data_df['Function'].replace([1000,2100,2110,2120,2130,2140,2150,2420,2490], 'Instruction')

finance_data_df['Function'] = finance_data_df['Function'].replace([2700,7100,7110,7120,7150,7180,7190,7191,7200,7210,7300,7310,
                                                                   7320,7330,7340,7350,7360,7370,7380,7390,7400,7410,7430,7490,
                                                                   7500,7510], 'Administration')

finance_data_df['Function'] = finance_data_df['Function'].replace([2495,4100,4200,4900,5000,5100,5400,5900,6000], 
                                                                  'Extra-Curricular/Community Activities')

finance_data_df['Function'] = finance_data_df['Function'].replace([7530,7540,7550,7600,7700,8100,8110,8200,8300,8400,8500,8700,
                                                                   9100,9200,9300], 'Operations')
finance_data_df
```

```python
finance_data_df['Function'].unique()
```

```python
# Calculate percentage of budgets
budget_percents = 100 * finance_data_df.groupby(["Function"]).sum()["Value"] / finance_data_df["Value"].sum()
budget_percents
```

```python
# Build percentage of funds pie chart
plt.subplots(figsize=(10,6))
plt.pie(budget_percents,
       labels= ["Administration", "Extra-Curricular/Community Activities", "Instruction", "Operations", "Student Services"],
       colors=["lightcoral", "orange", "gold", "lightskyblue", "lightgreen"],
       explode=[0,0,0.1,0,0],
       autopct='%1.1f%%',
       shadow=True, startangle=150)
plt.title("% of Funding by Category")

plt.show()
```
