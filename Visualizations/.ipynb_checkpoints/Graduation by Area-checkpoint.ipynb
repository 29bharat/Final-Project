```python
%matplotlib inline

# Import dependencies
import pandas as pd
import gmaps
import requests

# Import the API key
from config import g_key
```

```python
# Files to load
grad_1617_load = "./rate_1617.csv"
```

```python
# Store 16-17 Data into DataFrame
grad_1617_df = pd.read_csv(grad_1617_load)
grad_1617_df
```

```python
# Drop extra index column 
grad_1617_df = grad_1617_df.drop(["Unnamed: 0"], axis=1)
grad_1617_df
```

```python
grad_1617_df.dtypes
```

```python
# Load School district file
district_load = "../Resources/District Information Data/pubdistricts.csv"
```

```python
# Store District Data into data frame
district_df = pd.read_csv(district_load)
district_df.dtypes
```

```python
# Delete unnecessary columns
district_df = district_df.drop(['MailStreet', 'MailCity', 'MailZip', 'MailState', 'Phone ', 'Ext  ', 'AdmFName','AdmLName',
                                'AdmEmail', 'DOC','DOCType','StatusType','LastUpDate'], axis=1)
district_df
```

```python
# Drop last unnecessary row
district_df = district_df.drop([1103], axis=0)
district_df
```

```python
# Create columns for Dcode
district_df['Dcode'] = district_df['CD Code'].str[-5:]
district_df
```

```python
# Delete additional columns so that it is only Dcode, latitude, and longitude
district_df = district_df.drop(['CD Code','County', 'District', 'Street', 'City', 'Zip', 'State'], axis=1)
district_df
```

```python
# Rearrange columns 
district_df = district_df[['Dcode', 'Latitude', 'Longitude']]
district_df
```

```python
district_df.dtypes
```

```python
# Convert Dcode to numeric
district_df["Dcode"] = pd.to_numeric(district_df["Dcode"])
district_df.dtypes
```

```python
# Merge district_df and grad_1617_df
district_complete_df = pd.merge(district_df, grad_1617_df, on=["Dcode", "Dcode"])
district_complete_df
```

```python
# Configure Gmaps to use Google API key
gmaps.configure(api_key=g_key)
```

```python
# Create heatmap for Fund value

# 1. Assign the locations to an array of latitude and longitude pairs
locations = district_complete_df[["Latitude", "Longitude"]]

# 2. Assign the weights variable to some values
rate = district_complete_df["Grad_Rate"]

# 3. Assign the figure variable
fig = gmaps.figure(center=(37.2,-120), zoom_level=5.4)

# 4. Assign the Heatmap variables
heat_layer = gmaps.heatmap_layer(locations, weights=rate, dissipating=False, max_intensity=100,point_radius=.4)

# 5. Add the heatmap layer
fig.add_layer(heat_layer)

# 6. Call the figure to plot the data
fig
```
