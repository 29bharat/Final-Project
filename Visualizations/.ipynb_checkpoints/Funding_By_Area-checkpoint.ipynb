```python
%matplotlib inline

# Import dependencies
import pandas as pd
import gmaps
import requests

# Import the API key
from config import g_key
```

```python
# Files to load
finance_1617_load = "../Resources/Finance Data/School Budget Data '16-'17.csv"
```

```python
# Store 16-17 Data into DataFrame
finance_1617_df = pd.read_csv(finance_1617_load, 
                              names=['Ccode', 'Dcode', 'SchoolCode', 'Fiscalyear', 'Period','Colcode','Account','Fund',
                                     'Resource','Projectyear','Goal','Function','Object','Value'])
# Delete values with Function value 0
finance_1617_df = finance_1617_df[finance_1617_df.Function != 0]
finance_1617_df

finance_1617_df
```

```python
# Group by school districts and sum fund values
district_funds = finance_1617_df.groupby(["Dcode"]).sum()["Value"] / 1000000
district_funds
```

```python
# Store into DataFrame
district_funds_df = pd.DataFrame(district_funds).reset_index()
district_funds_df
```

```python
district_funds_df.dtypes
```

```python
# Load School district file
district_load = "../Resources/District Information Data/pubdistricts.csv"
```

```python
# Store District Data into data frame
district_df = pd.read_csv(district_load)
district_df
```

```python
# Delete unnecessary columns
district_df = district_df.drop(['MailStreet', 'MailCity', 'MailZip', 'MailState', 'Phone ', 'Ext  ', 'AdmFName','AdmLName',
                                'AdmEmail', 'DOC','DOCType','StatusType','LastUpDate'], axis=1)
district_df
```

```python
# Drop last unnecessary row
district_df = district_df.drop([1103], axis=0)
district_df
```

```python
# Create columns for Dcode
district_df['Dcode'] = district_df['CD Code'].str[-5:]
district_df
```

```python
# Delete additional columns so that it is only Dcode, latitude, and longitude
district_df = district_df.drop(['CD Code','County', 'District', 'Street', 'City', 'Zip', 'State'], axis=1)
district_df
```

```python
# Rearrange columns 
district_df = district_df[['Dcode', 'Latitude', 'Longitude']]
district_df
```

```python
district_df.dtypes
```

```python
# Convert Dcode to numeric
district_df["Dcode"] = pd.to_numeric(district_df["Dcode"])
district_df.dtypes
```

```python
# Merge district_df and district_funds_df
district_complete_df = pd.merge(district_df, district_funds_df, on=["Dcode", "Dcode"])
district_complete_df
```

```python
district_complete_df = district_complete_df[district_complete_df.Value >= 0]
district_complete_df
```

```python
# Configure Gmaps to use Google API key
gmaps.configure(api_key=g_key)
```

```python
# Create heatmap for Fund value

# 1. Assign the locations to an array of latitude and longitude pairs
locations = district_complete_df[["Latitude", "Longitude"]]

# 2. Assign the weights variable to some values
funding = district_complete_df["Value"]

# 3. Assign the figure variable
fig = gmaps.figure(center=(37.2,-120), zoom_level=5.4)

# 4. Assign the Heatmap variables
heat_layer = gmaps.heatmap_layer(locations, weights=funding, dissipating=False, max_intensity=100,point_radius=.25)

# 5. Add the heatmap layer
fig.add_layer(heat_layer)

# 6. Call the figure to plot the data
fig
```
